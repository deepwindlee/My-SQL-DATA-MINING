```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
file_name ='分组排行.xlsx'   #'
df =pd.read_excel(file_name)#('市区排名.xlsx')
df.columns
```

```python
#df
```

```python
df_need = df[['开始时间', '结束时间', '行政区', '街道名称', 'PM25本期', 
       'PM10本期', 'NO2本期',  
        'O3本期', 'SO2本期', 'AQI本期', '综合指数本期']]
```

```python
df_need
```

```python
rank_kind ='min'
PM25 = df_need[[ '行政区', '街道名称', 'PM25本期']]#.set_index('街道名称')
PM25['group_sort_pm25']   =PM25.groupby(['行政区'])['PM25本期'].rank(ascending=1,method=rank_kind)
PM25['group_sort_pm25'] =PM25['group_sort_pm25'].astype(int)
PM25['PM25_rank']=PM25['PM25本期'].rank(ascending=1,method=rank_kind)
PM25['PM25_rank'] =PM25['PM25_rank'].astype(int)
PM25[PM25['行政区']=='秦淮区']
```

```python
PM25
```

```python
rank_kind ='min'
PM10 = df_need[[ '行政区', '街道名称', 'PM10本期']]
PM10['group_sort_PM10']   =PM10.groupby(['行政区'])['PM10本期'].rank(ascending=1,method=rank_kind)
PM10['group_sort_PM10'] =PM10['group_sort_PM10'].astype(int)
PM10['PM10_rank']=PM10['PM10本期'].rank(ascending=1,method=rank_kind)
PM10['PM10_rank'] =PM10['PM10_rank'].astype(int)
PM10[PM10['行政区']=='秦淮区']
```

```python
rank_kind ='min'
NO2 = df_need[[ '行政区', '街道名称', 'NO2本期']]
NO2['group_sort_NO2']   = NO2.groupby(['行政区'])['NO2本期'].rank(ascending=1,method=rank_kind)
NO2['group_sort_NO2'] = NO2['group_sort_NO2'].astype(int)
NO2['NO2_rank']=NO2['NO2本期'].rank(ascending=1,method=rank_kind)
NO2['NO2_rank'] =NO2['NO2_rank'].astype(int)
NO2[NO2['行政区']=='秦淮区']
```

```python
rank_kind ='min'
O3 = df_need[[ '行政区', '街道名称', 'O3本期']]
O3['group_sort_O3']   = O3.groupby(['行政区'])['O3本期'].rank(ascending=1,method=rank_kind)
O3['group_sort_O3'] =O3['group_sort_O3'].astype(int)
O3['O3_rank']=O3['O3本期'].rank(ascending=1,method=rank_kind)
O3['O3_rank'] =O3['O3_rank'].astype(int)
O3[O3['行政区']=='秦淮区']
```

```python
R = PM25.merge(PM10).merge(O3).merge(NO2)
R
```

```python
R.to_excel('5月shang旬市排名和区排名.xlsx')
#[R['行政区']=='秦淮区'].to_excel('5月shang旬市排名和区排名.xlsx')
```
