## 导入相关包

```python
import pandas as pd 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文(windows)
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号
```

## 读取 农村数据，并重新命名列名

```python
import pandas as pd 
df_rural = pd.read_excel('农村.xlsx' ,skiprows=3)


df_rural.columns
df_rural.columns=['年份', '农村可支配收入', '指数(1978年=100)',
       '农村总支出', '生活消费支出', '食品支出', '每一劳动力负担人数', '人均居住面积']

df_rural.dropna(inplace=True)
df_rural
```

## 读取 城市数据，并重新命名列名

```python
df_urban = pd.read_excel('城市.xlsx' ,skiprows=3)
df_urban.columns=['年份', '城市全部收入','城市可支配收入' ,'指数(1978年=100)',
        '城市总支出', '食品支出', '每一劳动力负担人数', '人均居住面积']
df_urban.dropna(inplace=True)
df_urban
```

## 选取 城市和农村的 收入和 支出数据，并合并到一个dataframe中
新增 收入差  这一列

**以年份为索引**

```python
df_rual_urban = df_rural[['年份','农村可支配收入','农村总支出']].merge(df_urban[['年份','城市可支配收入','城市总支出']])
df_rual_urban =df_rual_urban.set_index('年份')
df_rual_urban['城市与农村收入差距'] = df_rual_urban['城市可支配收入']-df_rual_urban['农村可支配收入']
```

## 画出时序图
 ### 收入

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文(windows)
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号
# 图片大小设置
fig = plt.figure(figsize=(15,9), dpi=100)
ax1 = fig.add_subplot(111)
l1 =ax1.plot(df_rual_urban.index ,df_rual_urban['农村可支配收入'],color= 'r',label= '农村可支配收入')
l2 =ax1.plot(df_rual_urban.index ,df_rual_urban['城市可支配收入'],color= 'b',label= '城市可支配收入')
l22 =ax1.plot(df_rual_urban.index ,df_rual_urban['城市与农村收入差距'],color= 'grey',label= '收入差距')
ax1.set_ylabel('收入（元）', fontsize=15)
ax1.set_xlabel('年份', fontsize=18)
lns = l1 + l2 +l22 
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='center left', fontsize=18)
```

## 分析

用折线图然后做个对比，表明 :   
居民的年均收入对消费水平确实存在影响.，
消费与收入存在着一种明显的正相关关系,即随着收入的增加,消费相应增加,
反之,收入减少,消费相应减少。 
农村居民与城镇居民与农村居民人均收入差距很大

# 总结

本文通过对湖南省各个年份城市和农村收入与支出的数据的分析，使用Matplotlib、Pandas第三方库为工具。
得到以下结论：

1、随着年份的增加 对于湖南省，城市和农村的收入和支出都在增加，但是<font color=red> 农村和城市的差距也变大了。 </font> 

2、收入和支出明细呈现<font color=red>正相关 </font>     


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文(windows)
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号
# 图片大小设置
fig = plt.figure(figsize=(15,9), dpi=100)
ax2 = fig.add_subplot(111)
ax2.set_xlabel('年份', fontsize=18)
l3 =ax2.plot(df_rual_urban.index ,df_rual_urban['农村总支出'],color= 'y',label= '农村总支出')
l4 =ax2.plot(df_rual_urban.index ,df_rual_urban['城市总支出'],color= 'pink',label= '城市总支出')
#ax2.set_xlim(left=0, right=400)
#plt.xlabel('年份', fontsize=14)    # X轴标签
ax2.set_ylabel("支出（元）", fontsize=16) 
ax2.set_ylim(0, 7500)
lns =  l3 + l4
labs = [l.get_label() for l in lns]
plt.title("支出对比", fontsize=25, color='black', pad=20)
ax2.legend( loc='center left', fontsize=18)
plt.show()
```
