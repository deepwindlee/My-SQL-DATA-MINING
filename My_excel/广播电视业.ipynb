## 读取 住房指数  并重新命名  画图

```python
import pandas as pd 
df_house  = pd.read_excel('住房指数.xlsx' ,skiprows=0,sheet_name='Sheet1')
df_house.columns=['年份', '新建住宅', '新建商品住宅', '新建90 平米以下', '新建90-144平方米', '新建144平方米以上', '二手住宅',
       '二手90平米以下', '二手90-144平方米', '二手144平方米以上']
df_house
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文(windows)
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号
# 图片大小设置
fig = plt.figure(figsize=(15,9), dpi=100)
ax1 = fig.add_subplot(111)
l1 =ax1.plot(df_house.set_index('年份').index ,df_house['新建住宅'],color= 'r',label= '新建住宅')
l2 =ax1.plot(df_house.set_index('年份').index ,df_house['二手住宅'],color= 'b',label= '二手住宅')
l22 =ax1.plot(df_house.set_index('年份').index ,df_house['新建90 平米以下']
              ,color= 'y',label= '新建90平米以下')
l23 =ax1.plot(df_house.set_index('年份').index ,df_house['二手90平米以下']
              ,color= 'pink',label= '二手90平米以下')


ax1.set_ylabel('指数', fontsize=17)
ax1.set_xlabel('年份', fontsize=19)
lns = l1 + l2 +l22 +l23
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='center left', fontsize=18)
```

从上图可以看出， 湖南省总体房价趋于平稳， 特别是二手房 房价。

## 读取 各个城市设施水平   画图

```python
df_equip = pd.read_excel('设施水平.xlsx' ,skiprows=0)
df_equip
```

###  处理脏数据， 讲字符串转化为 浮点数方便画图

```python
col =['人口密度（人/平方公里）', '人均日生活用水量（升）', '用水普及率（%）', '每万人拥有公共交通车辆（台）',
       '用气普及率（%）', '人均拥有道路面积（平方米）', '排水管道密度（公里/平方公里）', '污水处理率（%）',
       '人均绿地面积（平方米）', '建成区绿地率（%）', '建成区绿地覆盖率（%）', '生活垃圾无害化处理率（%）']

for each in col:
    print('转化',each ,'为浮点数')
    df_equip[each] =  df_equip[each].astype(float)
```

```python
axes = df_equip.set_index('城市').plot.bar(rot=0, subplots=True ,figsize=(19,29))
axes[0].legend(loc=7)  
```

从上图可以看出， 长沙的 公共交通发达， 建成区绿化率高，人口密度适中

## 读取广播电视数据，并画图


```python
df_dianshi = pd.read_excel('广播电视.xlsx',sheet_name='Sheet1')
df_dianshi
```
